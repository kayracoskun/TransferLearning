# Modelin Eğitilmesi

Tensorflow/Keras kütüphanelerinin yüklenmesi

```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
# mobilenet modeli kullanıldı

mobilenet = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

pretrained_model = hub.KerasLayer(mobilenet, input_shape=(224, 224, 3), trainable=False)
```

```python
num_of_classes = 2 # cat & dog

model = tf.keras.Sequential([
    pretrained_model, 
    tf.keras.layers.Dense(num_of_classes) # output layer 
])

model.summary()
```

```python
model.compile(
    optimizer = 'adam', 
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(X_train_scaled, y_train, epochs=10)
```

10 epoch yerine 7-8 epoch yeterliymiş

```python
score, acc = model.evaluate(X_test_scaled, y_test)
print("Test loss: ", score)
print("Test accuraacy:", acc)
```
